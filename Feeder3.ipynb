# Looking at Parts of the Data Set

This notebook will instruct you on how to pull out only certain parts of a large data set to analyze.

### Step 1
First, you should download your data set as a .csv file

Next, import the numpy and pandas packages using the code below.

```python
import numpy as np
import pandas as pd
```

We need pandas to read the .csv file, so we would use pd.read_csv

Next we have to define pd.read_csv as df Use the following code below to do this

```python
df=pd.read_csv("CountyHealthData_2014-2015.csv")
```

### Step 2-Pick Your Topics

If you just want to find the correlation between Alchohol-impared deaths and excessive drinking you would need to locate those from the whole data set.


.loc will locate the parts of the data you want to see. We will use df.loc, parenthesis and then the titles of the parts of data you want to see. 
 
Use the following command to locate them. You can replace the words in parenthesis with two or more different titles.

```python
df.loc[:,["Alcohol-impaired driving deaths","Excessive drinking"]]
```

I used the same thing for the next set of data. I wanted to find the correlation between "Access to exercise opportunities" and "Adult obesity" from the whole data set.

You can use the following command again.

```python
df.loc[:,["Access to exercise opportunities","Adult obesity"]]
```

You can use the same command if you wanted to find just "children in poverty" by region. 

I used the following function to locate and show children in poverty in each region.

```python
df.loc[:,["Region", "Children in poverty"]]
```

```python
series0 = pd.DataFrame(df[["Region", "Children in poverty"]])
```

```python
series1=pd.DataFrame(df[["Access to exercise opportunities","Adult obesity"]])
```

```python
series2=pd.DataFrame(df[["Alcohol-impaired driving deaths","Excessive drinking"]])
```

```python
final = pd.concat([series0,series1,series2], axis=1,ignore_index=False, sort=False)
final
```

```python
final.to_csv("subdata.csv",index=False)
```
